```python
import os
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, StratifiedKFold, learning_curve
from sklearn.metrics import (
    classification_report, confusion_matrix, accuracy_score,
    RocCurveDisplay, roc_auc_score
)
from sklearn.preprocessing import OneHotEncoder, label_binarize
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.multiclass import OneVsRestClassifier
```

```python
PATH_UNLABELLED = "./dataset/preprocessed_unlabelled_data_bert.xlsx"
PATH_LABELLED  = "./dataset/processed_labeled_dataset_without_encoding.xlsx"
```

```python
TARGET_COL = "fraudulent"
TEXT_COLS = ["job_desc", "skills_desc", "text"]  # will be combined into 'all_text'
STRUCTURED_COLS = [
    "job_title", "location", "industry", "salary_range",
    "company_profile", "employment_type"
]
```

```python
TEST_SIZE = 0.2
RANDOM_STATE = 42
N_ESTIMATORS = 400
N_JOBS = -1
```

```python
USE_SIMPLE_OVERSAMPLE = True
```

```python
if not os.path.exists(PATH_LABELLED):
    raise FileNotFoundError(f"Labelled dataset not found: {PATH_LABELLED}")
if not os.path.exists(PATH_UNLABELLED):
    print(f"[Info] Unlabelled dataset not found: {PATH_UNLABELLED} (we can still train without it)")

df_labelled = pd.read_excel(PATH_LABELLED)
df_unlabelled = pd.read_excel(PATH_UNLABELLED) if os.path.exists(PATH_UNLABELLED) else None

print("Labelled shape:", df_labelled.shape)
display(df_labelled.head())

if df_unlabelled is not None:
    print("Unlabelled shape:", df_unlabelled.shape)
    display(df_unlabelled.head())
```

```python
# Ensure required columns exist in labelled data
needed_labelled = set(STRUCTURED_COLS + TEXT_COLS + [TARGET_COL])
missing_lab = needed_labelled - set(df_labelled.columns)
if missing_lab:
    raise ValueError(f"Missing in labelled data: {missing_lab}")

# Helper to combine text fields
def build_all_text(df, text_cols):
    def _join_row(row):
        parts = []
        for c in text_cols:
            val = row.get(c, "")
            if pd.isna(val):
                val = ""
            parts.append(str(val))
        return " ".join(parts)
    return df.apply(_join_row, axis=1)

df_labelled = df_labelled.copy()
df_labelled["all_text"] = build_all_text(df_labelled, TEXT_COLS)

if df_unlabelled is not None:
    # Accept that unlabelled may miss some columns; fill gracefully
    for c in STRUCTURED_COLS + TEXT_COLS:
        if c not in df_unlabelled.columns:
            df_unlabelled[c] = ""
    df_unlabelled = df_unlabelled.copy()
    df_unlabelled["all_text"] = build_all_text(df_unlabelled, TEXT_COLS)

display(df_labelled[STRUCTURED_COLS + [TARGET_COL, "all_text"]].head())
```

```python
y = df_labelled[TARGET_COL]
cls_counts = y.value_counts().sort_index()

plt.figure(figsize=(6,4))
cls_counts.plot(kind="bar")
plt.title("Class Distribution (Original)")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()

for cls, cnt in cls_counts.items():
    print(f"class {cls}: {cnt}")
```

```python
# === Clean & normalize target, then split safely ===

# 1) Normalize the target column to numeric (0/1 if it's binary); drop NaNs
def normalize_label(v):
    if pd.isna(v):
        return np.nan
    if isinstance(v, str):
        s = v.strip().lower()
        if s in {"1", "true", "t", "yes", "y"}:  return 1
        if s in {"0", "false", "f", "no", "n"}:   return 0
        # try numeric strings
        try:
            return int(float(s))
        except Exception:
            return np.nan
    # numeric-like
    try:
        return int(float(v))
    except Exception:
        return np.nan

df_labelled = df_labelled.copy()
df_labelled[TARGET_COL] = df_labelled[TARGET_COL].apply(normalize_label)

nan_before = df_labelled[TARGET_COL].isna().sum()
print(f"[Info] NaNs in `{TARGET_COL}` before drop: {nan_before}")

# Drop rows with NaN target
df_labelled = df_labelled[~df_labelled[TARGET_COL].isna()].copy()
df_labelled[TARGET_COL] = df_labelled[TARGET_COL].astype(int)

# 2) Build X, y (assumes you already created `all_text` in a prior cell)
X = df_labelled[STRUCTURED_COLS + ["all_text"]]
y = df_labelled[TARGET_COL]

# 3) Decide if stratify is safe (each class must have at least 1 sample in BOTH splits)
counts = y.value_counts().sort_index()
print("[Info] Class counts after cleaning:", counts.to_dict())

def can_stratify(class_counts, test_size):
    # need at least 2 total per class and at least 1 in each split
    for c, n in class_counts.items():
        n_test = int(np.floor(n * test_size))
        n_train = n - n_test
        if n < 2 or n_test < 1 or n_train < 1:
            return False
    return True

use_stratify = can_stratify(counts, TEST_SIZE)
print(f"[Info] Using stratify: {use_stratify}")

# 4) Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=TEST_SIZE,
    random_state=RANDOM_STATE,
    stratify=y if use_stratify else None
)

print("Train size:", X_train.shape, " Test size:", X_test.shape)

# 5) Optional: simple oversampling on TRAIN to balance classes
if USE_SIMPLE_OVERSAMPLE:
    counts_tr = y_train.value_counts()
    max_count = counts_tr.max()
    frames, labels = [], []
    for cls, cnt in counts_tr.items():
        idx = y_train[y_train == cls].index
        resample_idx = np.random.choice(idx, size=max_count, replace=True)
        frames.append(X_train.loc[resample_idx])
        labels.append(np.full(max_count, cls))
    X_train_bal = pd.concat(frames, axis=0)
    y_train_bal = pd.Series(np.concatenate(labels), name=TARGET_COL, index=X_train_bal.index)

    # Shuffle the balanced set
    perm = np.random.permutation(len(X_train_bal))
    X_train_bal = X_train_bal.iloc[perm]
    y_train_bal = y_train_bal.iloc[perm]
    print("[Info] Oversampling applied on train split.")
else:
    X_train_bal, y_train_bal = X_train, y_train

# 6) Plot the (balanced) train distribution
plt.figure(figsize=(6,4))
y_train_bal.value_counts().sort_index().plot(kind="bar")
plt.title("Train Class Distribution (After Simple Oversampling)" if USE_SIMPLE_OVERSAMPLE
          else "Train Class Distribution (No Oversampling)")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
# ColumnTransformer:
# - 'text': TF-IDF on 'all_text'
# - 'cat' : OneHot on structured columns (with imputation)
pre = ColumnTransformer(
    transformers=[
        ("text", TfidfVectorizer(max_features=70000, ngram_range=(1,2)), "all_text"),
        ("cat", Pipeline(steps=[
            ("imputer", SimpleImputer(strategy="most_frequent")),
            ("ohe", OneHotEncoder(handle_unknown="ignore"))
        ]), STRUCTURED_COLS)
    ],
    remainder="drop",
    verbose_feature_names_out=False
)

rf = RandomForestClassifier(
    n_estimators=N_ESTIMATORS,
    random_state=RANDOM_STATE,
    n_jobs=N_JOBS
)

pipe = Pipeline(steps=[
    ("pre", pre),
    ("rf", rf)
])

pipe
```

```python
pipe.fit(X_train_bal, y_train_bal)

y_pred = pipe.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print(f"Test Accuracy: {acc:.4f}\n")

print("Classification Report:")
print(classification_report(y_test, y_pred, zero_division=0))

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.metrics import confusion_matrix, roc_curve, auc, classification_report, accuracy_score
import pickle

# Set style for better plots
plt.style.use('default')
sns.set_palette("husl")

# 1. Class Distribution (from the existing plot)
def plot_class_distribution(y_train_bal, y_test):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
    
    # Train distribution
    train_counts = y_train_bal.value_counts().sort_index()
    ax1.bar(train_counts.index, train_counts.values)
    ax1.set_title('Train Class Distribution\n(After Oversampling)')
    ax1.set_xlabel('Class')
    ax1.set_ylabel('Count')
    ax1.set_xticks([0, 1])
    ax1.set_xticklabels(['0 (Non-fraud)', '1 (Fraud)'])
    
    # Add value labels on bars
    for i, v in enumerate(train_counts.values):
        ax1.text(i, v + 100, str(v), ha='center', va='bottom')
    
    # Test distribution
    test_counts = y_test.value_counts().sort_index()
    ax2.bar(test_counts.index, test_counts.values)
    ax2.set_title('Test Class Distribution')
    ax2.set_xlabel('Class')
    ax2.set_ylabel('Count')
    ax2.set_xticks([0, 1])
    ax2.set_xticklabels(['0 (Non-fraud)', '1 (Fraud)'])
    
    # Add value labels on bars
    for i, v in enumerate(test_counts.values):
        ax2.text(i, v + 50, str(v), ha='center', va='bottom')
    
    plt.tight_layout()
    plt.show()
    
    print(f"Train set - Class 0: {train_counts[0]}, Class 1: {train_counts[1]}")
    print(f"Test set - Class 0: {test_counts[0]}, Class 1: {test_counts[1]}")

# 2. Confusion Matrix
def plot_confusion_matrix(y_test, y_pred):
    cm = confusion_matrix(y_test, y_pred)
    
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=['Non-fraud (0)', 'Fraud (1)'], 
                yticklabels=['Non-fraud (0)', 'Fraud (1)'])
    plt.title('Confusion Matrix')
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show()
    
    return cm

# 3. ROC and AUC Curve
def plot_roc_auc(y_test, y_pred_proba):
    fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba[:, 1])
    roc_auc = auc(fpr, tpr)
    
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.4f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Classifier')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc="lower right")
    plt.grid(True, alpha=0.3)
    plt.show()
    
    return roc_auc

# 4. Learning Curve (Simulated since we don't have training history)
def plot_learning_curve(pipe, X_train_bal, y_train_bal, X_test, y_test):
    # For Random Forest, we can plot feature importance instead
    if hasattr(pipe.named_steps['rf'], 'feature_importances_'):
        # Get feature names from the preprocessor
        try:
            feature_names = pipe.named_steps['pre'].get_feature_names_out()
            importances = pipe.named_steps['rf'].feature_importances_
            
            # Get top 20 features
            indices = np.argsort(importances)[-20:]
            
            plt.figure(figsize=(10, 8))
            plt.title('Top 20 Feature Importances')
            plt.barh(range(len(indices)), importances[indices], color='b', align='center')
            plt.yticks(range(len(indices)), [feature_names[i] for i in indices])
            plt.xlabel('Relative Importance')
            plt.tight_layout()
            plt.show()
        except:
            # Fallback: general feature importance plot
            plt.figure(figsize=(10, 6))
            importances = pipe.named_steps['rf'].feature_importances_
            plt.bar(range(len(importances)), importances)
            plt.title('Feature Importances')
            plt.xlabel('Feature Index')
            plt.ylabel('Importance')
            plt.show()

# 5. Print accuracies and save model
def print_metrics_and_save(pipe, X_train_bal, y_train_bal, X_test, y_test, y_pred, y_pred_proba):
    train_accuracy = accuracy_score(y_train_bal, pipe.predict(X_train_bal))
    test_accuracy = accuracy_score(y_test, y_pred)
    
    print("="*50)
    print("MODEL PERFORMANCE METRICS")
    print("="*50)
    print(f"Train Accuracy: {train_accuracy:.4f}")
    print(f"Test Accuracy:  {test_accuracy:.4f}")
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred, target_names=['Non-fraud (0)', 'Fraud (1)']))
    
    # Calculate additional metrics
    from sklearn.metrics import precision_score, recall_score, f1_score
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    
    print(f"\nAdditional Metrics:")
    print(f"Precision: {precision:.4f}")
    print(f"Recall:    {recall:.4f}")
    print(f"F1-Score:  {f1:.4f}")
    
    # Save model as .pkl
    with open('fraud_detection_model.pkl', 'wb') as f:
        pickle.dump(pipe, f)
    print(f"\nModel saved as 'fraud_detection_model.pkl'")
    
    # For .keras format, we would need a Keras model
    # Since this is scikit-learn, we'll save as joblib which is better for sklearn models
    import joblib
    joblib.dump(pipe, 'fraud_detection_model.joblib')
    print(f"Model saved as 'fraud_detection_model.joblib'")

# Main execution function
def generate_all_plots_and_metrics(pipe, X_train_bal, y_train_bal, X_test, y_test):
    # Get predictions
    y_pred = pipe.predict(X_test)
    y_pred_proba = pipe.predict_proba(X_test)
    
    # 1. Class Distribution
    print("1. CLASS DISTRIBUTION")
    plot_class_distribution(y_train_bal, y_test)
    
    # 2. Confusion Matrix
    print("2. CONFUSION MATRIX")
    cm = plot_confusion_matrix(y_test, y_pred)
    
    # 3. ROC and AUC
    print("3. ROC AND AUC CURVE")
    roc_auc = plot_roc_auc(y_test, y_pred_proba)
    print(f"AUC Score: {roc_auc:.4f}")
    
    # 4. Learning Curve / Feature Importance
    print("4. FEATURE IMPORTANCE")
    plot_learning_curve(pipe, X_train_bal, y_train_bal, X_test, y_test)
    
    # 5. Metrics and Save
    print_metrics_and_save(pipe, X_train_bal, y_train_bal, X_test, y_test, y_pred, y_pred_proba)
    
    return {
        'confusion_matrix': cm,
        'auc_score': roc_auc,
        'y_pred': y_pred,
        'y_pred_proba': y_pred_proba
    }

# Execute the analysis
results = generate_all_plots_and_metrics(pipe, X_train_bal, y_train_bal, X_test, y_test)

# Additional: Save results dictionary
with open('model_results.pkl', 'wb') as f:
    pickle.dump(results, f)
print("Results dictionary saved as 'model_results.pkl'")
```
